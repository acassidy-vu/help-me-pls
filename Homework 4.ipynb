```python
#HW4_Patient_FBG has 3 fasting blood glucose (FBG) measures across 3 years for 200 patients. 
#HW4_Patient_Sex has the same set of patients and their reported Sex

#'HW4_Patient_FBG' is almost identical to the FBG file you all used in HW3 with one important 
#difference; There are now missing values in the file (denoted as 'NA').

#With these files, we want you to determine the diabetic status for each patient and write out how 
#many of each status there are and break it down by Sex. 
#There are three possible statuses: Healthy, Diabetic, and NA. 
#Once again we would like you to write out your answer to either a '.txt' file or a '.csv' file. 
#Your output should look like something like this (though feel free to format it however you want):
    #Sex     Healthy  Diabetic   NA
    #Male    34       24         4
    #Female  43       41         1
    
#Rules for determining patient status:
    #A single fasting blood glucose measure of 125 and below is considered a healthy measure
    #A single fasting blood glucose measure of 126 and above is considered a diabetic measure
    #If a subject has only healthy measures, they are considered healthy
    #If a subject has ANY diabetic measures, they are considered diabetic
    #If a subject has no measures (only 'NA'), then they go under the 'NA' category

#Notes and tips:
    #-The Patient_Sex.csv file is not ordered by the patient number
    #-I recommend that your code reads through the Patient_sex file first and that you store that 
        #information somewhere (perhaps a Dictionary or a list  might be useful...)
    #-You do not need to send to us the diabetic status for each patient; All we need to is the total
        # #s for each category(see example answer)
    #-Your code will have to check if a value is 'NA' or not, an easy way to do this is to use the 
        #".isnumeric()" function. This function returns true if your value is a number (even if it's 
        #technically a string) and false if the value is not a number. A simple if statement checking 
        #if the value is 'NA' would also work just fine.
    #-Making a function which accepts a series of fbg values and returns diabetic status could make 
    #things easier for you, but feel free to approach this problem however you want
```

```python
#***************************************ALL CODE REPRESENTS MY OWN WORK**************************************************


#------------------------STEP 1: OPEN PATIENT SEX AND CREATE A DICTIONARY OF THOSE VALUES---------------------------------
pat_sx={}  #this is an empty dictionary that I will add values to

sex=open(r'C:\Users\annao\Desktop\HW4_Patient_Sex.csv','r')
pt_sex=sex.readlines()
pt_sex.pop(0)
for items in pt_sex:
    sx=items.strip().split(sep=',') #I have a new list called sx with all the patient IDs and their corresponding sex
    #I want to create a dictionary so I need to define the key and the value
    key=sx[0]       #I am using sx[0] since the key is the patient ID which is the first item each sx list
    value=sx[1]     #the value is the sex
    pat_sx[key]=value #I am adding these values from sx to the dictionary called pat_sx 
#print(pat_sx)
sex.close()

#------------------------STEP 2: OPEN PATIENT FBG AND DETERMINE PATIENT STATUS-------------------------------------------------
    #If one of the values in greater than 126 then I want to lookup the patient in the dictionary and
        #check sex, then I want to characterize as diabetic and increase the number of diabetic 
        #individuals for the appropriate sex
    #If all the values are NA, I want to lookup the patient in the dictionary and
        #check sex, then I want to characterize as NA and increase the number of NA
        #individuals for the appropriate sex  
    #If all of the values are <=125 then I want to lookup the patient in the dictionary and
        #check sex, then I want to characterize as healthy and increase the number of healthy 
        #individuals for the appropriate sex

        
#each of these values will be used to count the number of individuals with their corresponding status
# F=female, M=male
# H=healthy, D=diabetic, NA=NA
FH=0
FD=0
FNA=0
MH=0
MD=0
MNA=0


fbg=open(r'C:\Users\annao\Desktop\HW4_Patient_FBG.csv','r')
pt_fbg=fbg.readlines()
pt_fbg.pop(0)
for itms in pt_fbg:
    fg=itms.strip().split(sep=',')
    
    #error handling, change all NA values to 0
    for val in range(len(fg[1:])):  #using len since I will be using index to change NA to 0
        if fg[val+1]=='NA': #using +1 since I started at 1, so I must shift the index
            fg[val+1]=0 #assinging NA to be equal to zero
        else:
            fg[val+1]=int(fg[val+1]) #I am turning the other numerical values into an integer so that I can 
                                        #execute the code below
    
    patient_ID=fg[0]  #setting patient ID so that I can use this later to check the sex in my sex dictionary
    if fg[1]>=126 or fg[2]>=126 or fg[3]>= 126:    #want to designate as diabetic
                                  #check sex of appropriate patient based on 
        if pat_sx[patient_ID]=='F':
            FD +=1  #increase the count of the FD value, will similarly repeat for all others 
        elif pat_sx[patient_ID]=='M':
            MD +=1
   
    elif fg[1]==0 and fg[2]==0 and fg[3]==0: #must be all values equal to zero
        #want to designate as NA
          #check sex of appropriate patient based on 
        if pat_sx[patient_ID]=='F':
            FNA +=1
        elif pat_sx[patient_ID]=='M':
            MNA +=1
            
    else:  #want to designate as healthy
          #check sex of appropriate patient based on 
        if pat_sx[patient_ID]=='F':
            FH +=1
        elif pat_sx[patient_ID]=='M':
            MH +=1

fbg.close()
    
#-------------------STEP 3: CREATE A NEW DOCUMENT AND WRITE THE RESULTS FROM THE CODE ABOVE INTO IT -----------------------

slt=open(r'C:\Users\annao\Desktop\Wk_4_Sex_FBG.txt','w')                   #this is my new file 
slt.write('Sex       '+'Healthy   '+'Diabetic  '+'NA        '+'\n')
slt.write('Female    '+ str(FH)+'        '+ str(FD)+ '        '+ str(FNA)   +'\n')
slt.write('Male      '+ str(MH)+'        '+ str(MD)+ '        '+ str(MNA)   )
slt.close()

slt=open(r'C:\Users\annao\Desktop\Wk_4_Sex_FBG.txt','r')  #I am also reading the values and printing them here
print(slt.read()) 
slt.close()
  
```
